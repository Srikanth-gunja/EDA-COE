### 1 . Handling Missing Values

```python
# Check Null Values
df.isnull().sum()
```

```python
c_df=df.drop_duplicates() # Droping Duplicates
```

```python
c_df.shape
```

```python
# Filling Missing Values
```

```python
c_df['Title'].unique()
```

```python
# print(c_df.iloc[2,0])
```

```python
c_df['Title'].fillna(c_df['Title'].mode()[0],inplace=True)
```

```python
c_df['Company'].fillna(c_df['Company'].mode()[0],inplace=True)
```

```python
cat=c_df.select_dtypes(exclude='number').columns
cat
```

```python
for col in cat:
    c_df[col].fillna(c_df[col].mode()[0],inplace=True)
```

```python
c_df.isnull().sum()
```

```python
c_df.drop(columns=['Attach','Notes'],inplace=True) # Droping Attach and Notes
```

## 2 .String Operations & Text Processing

```python
c_df['JobDescription'][2] # Checking the Pattern to Remove
```

```python
# Removing \r\n-\ chars
c_df['JobDescription']=c_df['JobDescription'].str.replace(r'[\r\n/-]',"",regex=True).str.strip().str.capitalize()
```

```python
c_df['JobRequirment']=c_df['JobRequirment'].str.replace(r'[\r\n/-]',"",regex=True).str.strip().str.capitalize()
```

```python
list(c_df['Title'].unique())
# for i in n:
#     print(n)
```

## 3. Date Operations

```python
c_df['date']=pd.to_datetime(c_df['date'],format='mixed',errors='coerce')
```

```python
c_df['StartDate']=pd.to_datetime(c_df['StartDate'],errors='coerce')
```

```python
c_df['OpeningDate']=pd.to_datetime(c_df['OpeningDate'],errors='coerce')
```

```python
c_df['Deadline']=pd.to_datetime(c_df['Deadline'],errors='coerce')
```

```python
c_df.info()
```

```python
c_df.isna().sum()
```

```python
date_cols=['date','StartDate','OpeningDate','Deadline']
for i in date_cols:
    c_df[i]=c_df[i].fillna(c_df[i].mode()[0])
```

```python
c_df.info()
```

```python
c_df.isnull().sum()
```

```python
c_df['Deadline'].dt-c_df['OpeningDate'].dt
```

```python
## Calculating the Days left to apply
c_df['Days_left']=(c_df['Deadline']-c_df['OpeningDate']).dt.days
```

```python
c_df['Job_Delay']=(c_df['StartDate']-c_df['OpeningDate']).dt.days
```
